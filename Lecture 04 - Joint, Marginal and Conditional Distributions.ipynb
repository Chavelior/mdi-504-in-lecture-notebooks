```python
# for interactive plots
%matplotlib notebook

import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
# for 3D plots
from mpl_toolkits.mplot3d import Axes3D
# for color plots
from matplotlib import cm


x = np.linspace(400, 800, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)

mu_X = 600
mu_Y = 5

sigma_X = 75
sigma_Y = 2
sigma_XY = -100

Z = mlab.bivariate_normal(X, Y, sigma_X, sigma_Y, mu_X, mu_Y, sigma_XY)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.viridis, linewidth=0)

plt.figure()
plt.contour(X, Y, Z)
```

```python
plt.figure()
plt.plot(x,mlab.normpdf(x, mu_X, sigma_X), linewidth=3)
plt.grid()
```
